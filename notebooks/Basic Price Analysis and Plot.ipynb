```python
import matplotlib.pyplot as plt
import pandas as pd

df_asset = pd.read_parquet(f'../data/sectors/QQQ_daily.parquet')

# Calculate moving averages
df_asset['ma50'] = df_asset['close'].rolling(window=50).mean()
df_asset['ma200'] = df_asset['close'].rolling(window=200).mean()


plt.figure(figsize=(12, 4))
plt.plot(df_asset.index, df_asset['close'], 'b-', label='QQQ');
plt.plot(df_asset.index, df_asset['ma50'], 'r-', label='ma50');
plt.plot(df_asset.index, df_asset['ma200'], 'g-', label='ma200');
plt.title('QQQ')
plt.xlabel('Date')
plt.ylabel('Price [USD]')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def display_data_info(df):
    print("\nDataFrame Info:")
    print("-" * 50)
    print(df.info())
        
    print("\nBasic statistics:")
    print("-" * 50)
    print(df.describe())
    
    print("\nDate range:")
    print("-" * 50)
    print(f"Start date: {df.index.min()}")
    print(f"End date: {df.index.max()}")
    print(f"Total days: {len(df)}")

display_data_info(df_asset)
```
